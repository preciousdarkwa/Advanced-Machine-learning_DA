## Dataset Description

The Adidas Sales dataset includes the following columns:

- Retailer: Represents the name of the retailer involved in the sales transaction.
- Retailer_ID: A numeric identifier assigned to each retailer for tracking and identification purposes.
- Invoice_Date: The date when the sales transaction took place.
- Region: Denotes the broader geographical region where the sales activity occurred.
- State: Specifies the state or province associated with the sales transaction.
- City: Indicates the city where the sales took place.
- Product: Describes the name or identifier of the product being sold.
- Price_per_Unit: Represents the unit price of the product sold.
- Units_Sold: Signifies the quantity of units sold for a particular product in a single transaction.
- Total_Sales: The total monetary value generated by the sales transaction, calculated by multiplying the price per unit and units sold.
- Operating_Profit: Refers to the profit earned from the sales transaction.
- Sales_Method: Indicates the method or channel through which the sales transaction was conducted, such as online, in-store, or through a third-party platform.

## Data Cleaning

```python
#Renaming columns
adidas_sales.rename(columns={"Retail": "Retailer_ID",
            "Invoice Date": "Invoice_Date",
            "Price per Unit": "Price_per_Unit",
            "Units Sold": "Units_Sold",
            "Total Sales": "Total_Sales",
            "Operating Profit": "Profit",
            "Sales Method": "Sales_Method"}, inplace=True)
```

### Changing DataTypes

```python
adidas_sales.dtypes
```

```python
adidas_sales['Retailer']= adidas_sales['Retailer'].astype('category')
adidas_sales['Region']= adidas_sales['Region'].astype('category')
adidas_sales['Product']= adidas_sales['Product'].astype('category')
adidas_sales['Sales_Method']= adidas_sales['Sales_Method'].astype('category')
adidas_sales['Price_per_Unit'] = adidas_sales['Price_per_Unit'].astype(float)
adidas_sales['Units_Sold'] = adidas_sales['Units_Sold'].astype(int)
adidas_sales['Profit'] = adidas_sales['Profit'].astype(int)
adidas_sales['Invoice_Date'] = pd.to_datetime(adidas_sales['Invoice_Date'])
```

## Sales Analysis

### 1. Sales Performance by Retailer

```python
# Group the data by the retailers and sum total sales
grouped_data = adidas_sales.groupby('Retailer')['Total_Sales'].sum()

# Bar plot
plt.figure(figsize=(10, 5))
bar_plot = plt.bar(grouped_data.index, grouped_data.values)

# Add data labels
for bar in bar_plot:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height):,}', ha='center', va='bottom', color='black')

# Remove y-axis
plt.gca().yaxis.set_visible(False)

# Set the title and labels
plt.xlabel('Retailer')
plt.ylabel('Total Sales')
plt.title('Sales Performance by Retailer')

# Rotating the x-axis labels for better readability (optional)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Find the retailer with the highest sales
max_retailer = grouped_data.idxmax()
print(f"The retailer with the highest sales is: '{max_retailer}'")
```

### 2. Sales Distribution by Product Category

```python
# Grouping the data by product category and calculate the total sales for each category
category_sales = adidas_sales.groupby('Product')['Total_Sales'].sum()

# Pie chart
plt.figure(figsize=(10,5))
plt.pie(category_sales.values, labels=category_sales.index, autopct='%1.1f%%', shadow=True)

plt.title('Sales Distribution by Product Category')

plt.show()
```

### 3. Seasonal Sales Analysis

```python
# Group by 'Season' and sum 'Total Sales'
season_sales = adidas_sales.groupby('Season')['Total_Sales'].sum().reset_index()

# Create bar chart with adjusted figure size using Seaborn
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
bar_plot = sns.barplot(x='Season', y='Total_Sales', data=season_sales, palette='tab20b')

# Add data labels
for bar in bar_plot.patches:
    height = bar.get_height()
    bar_plot.text(bar.get_x() + bar.get_width() / 2, height, f'${int(height):,}', ha='center', va='bottom', color='white')

# Remove y-axis
plt.gca().yaxis.set_visible(False)

# Set the title and labels
plt.title('Seasonal Sales')
plt.xlabel('Season')
plt.ylabel('Total Sales')

# Save the plot as an image
plt.savefig('seasonal_sales.png')

# Display the plot
plt.show()

# Find the season with the highest sales
max_season = season_sales.loc[season_sales['Total_Sales'].idxmax(), 'Season']
print(f"The highest sales occurred in the Season: '{max_season}'")
```

### 4. Top 5 States by Sales

```python
# Grouping data and selecting top 5 states by total sales
grouped_data = adidas_sales.groupby('State')['Total_Sales'].sum().reset_index().nlargest(5, 'Total_Sales')

# Barplot
plt.figure(figsize=(10, 5))
bar_plot = sns.barplot(data=grouped_data, x='State', y='Total_Sales')

# Add data labels
for bar in bar_plot.patches:
    height = bar.get_height()
    bar_plot.text(bar.get_x() + bar.get_width() / 2, height, f'${int(height):,}', ha='center', va='bottom', color='black')

# Remove y-axis
plt.gca().yaxis.set_visible(False)

# Set the title and labels
plt.xlabel('States')
plt.ylabel('Total Sales')
plt.title('Top 5 States by Sales')

plt.show()
```

### 6. Sales Trend

```python
# Filter data for 2020 and 2021
adidas_sales_2020 = adidas_sales[adidas_sales['Year'] == 2020]
adidas_sales_2021 = adidas_sales[adidas_sales['Year'] == 2021]

# Group by month and sum 'Total Sales' for each year
sales_2020 = adidas_sales_2020.groupby('Month')['Total_Sales'].sum()
sales_2021 = adidas_sales_2021.groupby('Month')['Total_Sales'].sum()

# Plot trend chart
plt.figure(figsize=(10, 2.5))
plt.plot(sales_2020.index, sales_2020.values, label='2020', marker='+')
plt.plot(sales_2021.index, sales_2021.values, label='2021', marker='o')
plt.title('Total Sales Trend Comparison (2020 vs 2021)')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(False)

# Save the plot as an image
plt.savefig('sales_trend.png')

plt.show()
```

### Graphical Representation of Product Growth Rate

```python
# Separate sales data by year
adidas_sales_2020 = adidas_sales[adidas_sales['Year'] == 2020]
adidas_sales_2021 = adidas_sales[adidas_sales['Year'] == 2021]

# Group by Retailer and sum 'Total_Sales' for each year
sales_2020 = adidas_sales_2020.groupby('Product')['Total_Sales'].sum().reset_index()
sales_2021 = adidas_sales_2021.groupby('Product')['Total_Sales'].sum().reset_index()

# Rename columns for merging
sales_2020.columns = ['Product', 'sales_2020']
sales_2021.columns = ['Product', 'sales_2021']

# Merge sales data for 2020 and 2021
merged_sales = pd.merge(sales_2020, sales_2021, on='Product')

# Calculate the growth rate
merged_sales['Growth_Rate'] = (merged_sales['sales_2021'] - merged_sales['sales_2020']) / merged_sales['sales_2020'] * 100

# Sort values by growth rate
merged_sales = merged_sales.sort_values(by='Growth_Rate', ascending=False)

# Create horizontal bar chart with Seaborn using 'tab10' palette
plt.figure(figsize=(10, 5))
bar_plot = sns.barplot(x='Growth_Rate', y='Product', data=merged_sales, palette='tab10')

for bar in bar_plot.patches:
    width = bar.get_width()
    bar_plot.text(width, bar.get_y() + bar.get_height() / 2, f'{width:.2f}%', ha='center', va='center', color='black')

# Hide x-axis label
plt.gca().axes.get_xaxis().set_visible(True)  # Keep the y-axis visible for retailer names

# Set the title and labels
plt.title("Growth Rate by Product")
plt.xlabel("Growth Rate (%)")
plt.ylabel("")

# Save the plot as an image
plt.savefig('growth_rate.png')

# Display the plot
plt.show()
```
